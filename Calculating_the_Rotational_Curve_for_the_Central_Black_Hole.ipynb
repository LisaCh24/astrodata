```python
import astropy.units as u # import a Python package to call certain units.
import astropy.constants as const # import a Python package to call the value of certain constants.
import numpy as np  # import numpy package, which allows us to conduct mathematical operations for the defined quantities.
```

```python
def CalculatingOrbitalVelocity(M, R):
    """
    This function calculates the orbital velocity of an object.
    Input:
    M, the mass of central black hole, Sagittarius A.
    R, the orbital radius of the arbitary points
    Output:
    orbitalVelocity, the orbital velocity of the satellite
    """
    orbitalVelocity = np.sqrt(const.G * M / R)
    return orbitalVelocity # define what value the definition will give.
```

```python
# Below is an array of arbitrary radii in kpc in our solar system.
R_arr = (np.arange(1,30) * u.kpc)
print(R_arr.to(u.km))
```

```python
# Below we define input M.
M = 4100000 * u.solMass
```

```python
# Calculate the orbital velocities of the array.
v = CalculatingOrbitalVelocity(M, R_arr) # use the pre-defined function.
print(v.to(u.kpc/u.s)) # print the orbital velocity of the arbitary distances in kpc.
```

```python
# Use Gemini AI to help create a plot of the rotational curve.
# prompt: Can you help me with plotting v and R_arr?

import matplotlib.pyplot as plt # import plotting package.

# Plot the orbital velocities against the orbital radii.
plt.plot(R_arr.to(u.kpc), v.to(u.kpc/u.s)) # plot velocity against orbital radii.
plt.xlabel("Orbital Radius (kpc)") # add x-axis label.
plt.ylabel("Orbital Velocity (kpc/s)") #add y-axis
plt.title("Orbital Velocity vs. Radius") # title the plot.
plt.show() # show the plot below.
```
